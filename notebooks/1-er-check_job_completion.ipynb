# Check Repackaging Jobs completion

This very small notebook provide a simple way to find if the repacking jobs is finished by parsing the log file 

```python
import re
from pathlib import Path

import pandas as pd
```

```python
# Update this path to your log file
log_path = Path("../logs/repackage_calibration_results.log")

pattern = re.compile(r"in\s+(\d+\.\d+)\s+seconds\.")

times = []
with log_path.open("r", encoding="utf-8") as f:
    for line in f:
        m = pattern.search(line)
        if m:
            times.append(float(m.group(1)))

s = pd.Series(times, name="seconds")
stats = s.describe(percentiles=[0.25, 0.5, 0.75])
stats["median"] = s.median()
stats["std"] = s.std()
stats["var"] = s.var()
stats
```

```python
completiont_couter = 0
for line in log_path.open("r", encoding="utf-8"):
    if "Repackaged output already exists"  in line  or "Postprocessing completed" in line:
        completiont_couter += 1

print(f"Number of completed repackaging processes: {completiont_couter}/119")
```
